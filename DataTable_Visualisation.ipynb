```python
import pandas as pd 
import seaborn as sns #visualisation
```

```python
#Loading the CSV file into a pandas dataframe.
df = pd.read_csv('gapminder_data.csv')
df.head(2)
```

```python
# Create the default pairplot
sns.pairplot(df)
```

```python
#coloring the figures based on a categorical variable such as continent
sns.pairplot(df, hue = 'continent')
```

```python
#A better method for showing univariate (single variable) distributions from multiple categories is the density plot. 
# We can exchange the histogram for a density plot in the function call
# Create a pair plot colored byThe density plots on the diagonal make it easier 
# to compare distributions between the continents than stacked bars.  continent with a density plot of the 
# diagonal and format the scatter plots.
#The density plots on the diagonal make it easier to compare distributions 
# between the continents than stacked bars. 
sns.pairplot(df, hue = 'continent', diag_kind = 'kde',
             plot_kws = {'alpha': 0.6, 's': 80, 'edgecolor': 'k'},
             size = 4)
```

```python
#reduce the clutter by plotting only the years after 2000. We will still 
# color by continent, but now we won’t plot the year column. To limit the columns plotted, we pass in a list of vars to the function. To clarify the plot, we can also add a title.

# Plot colored by continent for years 2000-2007
sns.pairplot(df[df['year'] >= 2000], 
             vars = ['lifeExp', 'pop', 'gdpPercap'], 
             hue = 'continent', diag_kind = 'kde', 
             plot_kws = {'alpha': 0.6, 's': 80, 'edgecolor': 'k'},
             size = 4);
# Title 
#plt.suptitle('Pair Plot of Socioeconomic Data for 2000-2007', 
#             size = 28);
```
